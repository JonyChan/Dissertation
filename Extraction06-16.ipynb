```python
def getfun(path,num):
    img = cv.imread(path)
    rows, columns, channels = img.shape
    
    b_list = []
    g_list = []
    r_list = []
    
    for i in np.arange(rows):
        for j in np.arange(columns):
            b,g,r = img[i,j]
            b_list.append(b)
            g_list.append(g)
            r_list.append(r)
            
    b_left = []
    g_left = []
    r_left = []
    
    for i in range(0, len(b_list), num):
        b_left.append(random.sample(b_list[i:i + num], 1))
    #     print(i)

    for i in range(0, len(g_list), num):
        g_left.append(random.sample(g_list[i:i + num], 1))

    for i in range(0, len(r_list), num):
        r_left.append(random.sample(r_list[i:i + num], 1))

    while [255] in b_left:
        a = b_left.index([255])
        del b_left[a]
        del g_left[a]
        del r_left[a]

    while [255] in g_left:
        a = g_left.index([255])
        del b_left[a]
        del g_left[a]
        del r_left[a]

    while [255] in r_left:
        a = r_left.index([255])
        del b_left[a]
        del g_left[a]
        del r_left[a]
    
```
